```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import string
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
fuzzy_csv = 'fuzzy_df.csv'
billboard_csv = '../Resources/billboard_lyrics_1964-2015.csv'
song_data_csv = '../Resources/song_data.csv'
```

```python
fuzzy_df = pd.read_csv(fuzzy_csv)
billboard_df = pd.read_csv(billboard_csv, encoding = 'ISO-8859-1')
song_data_df = pd.read_csv(song_data_csv)
```

```python
song_data_df.columns
```

```python
# adding Song List Index column to merge with the fuzzy_df
song_data_df.insert(0,'Song List Index',range(len(song_data_df)))
```

```python
song_data_df.head()
```

```python
# merging
matched_df = pd.merge(fuzzy_df,song_data_df,on='Song List Index',left_index=True)
```

```python
matched_df.columns
```

```python
matched_df.rename(columns={'Unnamed: 0':'order'})
```

```python
matched_df.columns=matched_df.columns.str.strip()
```

```python
matched_df = matched_df.sort_values('order')
matched_df
```
